```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x = data.drop("phishing",axis=1)
y = data['phishing']
scaled_data = scale.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte = train_test_split(x,y,test_size=0.25,random_state=100)
```

```python
from sklearn.metrics import f1_score, confusion_matrix
```

```python
def check_model(model):
    model.fit(xtr,ytr)
    y_pre1 = model.predict(xtr)
    y_pre2 = model.predict(xte)
    score2 = f1_score(y_true=yte,y_pred=y_pre2)
    score1 = f1_score(y_true=ytr,y_pred=y_pre1)
    conf_score = confusion_matrix(y_true=yte,y_pred=y_pre2)
    print(f"Score of model: {model} on train dataset is {score1}")    
    print(f"Score of model: {model} on test dataset is {score2}")
    print(f"confusion metris score is {conf_score}")
```

```python
## Logistic regression
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
check_model(log_model)
```

```python
## Decision tree
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier()
check_model(tree_model)
```

```python
## Random Forest 
from sklearn.ensemble import RandomForestClassifier
forest_model = RandomForestClassifier()
check_model(forest_model)
```

```python
## Here we are getting good accuracy with Random forest with little bit of overfitting
## Lets hypertune the Random forrest model little bit
param = {
        'n_estimators':[50,70,100,120],
    'min_samples_split':[2,4,6,8],
    "min_samples_leaf":[1,2,4,6]  
}
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(forest_model,param)
grid.fit(xte,yte)
```

```python
grid.best_score_
```

```python
check_model(grid.best_estimator_)
```

```python
grid.best_params_
```

```python
upgrade_forest = RandomForestClassifier(n_estimators=120,bootstrap=True,class_weight="balanced",oob_score=True)
check_model(upgrade_forest)
```
